- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
homeworkdata = pd.read_csv('/content/gdrive/MyDrive/YBIGTA/Numpy Pandas/세션자료/deliver.csv')
homeworkdata.head()
```

```python
#month를 먼저 꺼내자
homeworkdata['월'] = pd.DatetimeIndex(homeworkdata['날짜']).month
homeworkdata.head()
```

```python
#계절을 추출하는 함수를 만들자
def seasons(val):
  if 3 <= val < 6:
    return '봄'
  
  elif 6 <= val < 9:
    return '여름'
  
  elif 8<= val < 12:
    return '가을'
  
  else:
    return '겨울'
  
homeworkdata['계절'] = homeworkdata['월'].apply(seasons)

homeworkdata.head()
```

```python
homeworkdata['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
homeworkdata['미세먼지(㎍/㎥)'].mean()
```

```python
homeworkdata['미세먼지(㎍/㎥)'] = homeworkdata['미세먼지(㎍/㎥)'].fillna(homeworkdata['미세먼지(㎍/㎥)'].mean())

homeworkdata.head(30)
```
